```python
import torch
```

```python
words = open("names.txt").read().splitlines()

n_words = len(words)
```

```python
next_word_freq = {}

for w in words:
    chs = ["<S>"] + list(w) + ["</S>"]
    for ch1, ch2 in zip(chs, chs[1:]):
        k =(ch1,ch2)
        next_word_freq[k] = next_word_freq.get(k, 0) + 1


# sorted(next_word_freq, lambda x: len(x))
print(next_word_freq)
```
