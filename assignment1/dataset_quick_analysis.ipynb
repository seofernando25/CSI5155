```python
import numpy as np
from pipeline import load_dataframe, TARGET_COL

df = load_dataframe()
df.head()
```

```python
counts = df[TARGET_COL].value_counts()
pcts = df[TARGET_COL].value_counts(normalize=True) * 100

print("Target distribution:")
for cls, cnt in counts.items():
    print(f"  {cls}: {cnt} ({pcts[cls]:.2f}%)")

imbalance = counts.max() / counts.min()
print(f"Imbalance ratio: {imbalance:.2f}:1")
```

```python
missing = df.isnull().sum()
missing = missing[missing > 0]

print("Missing values:")
for col, cnt in missing.items():
    pct = (cnt / len(df)) * 100
    print(f"  {col}: {cnt} ({pct:.2f}%)")
```

```python
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
print(f"Numeric features: {len(num_cols)}")
```

```python
cat_cols = df.select_dtypes(include=['object']).columns.tolist()
print(f"Categorical features: {len(cat_cols)}")

for name in cat_cols:
    unique_count = df[name].nunique(dropna=True)
    print(f"{name}: {unique_count} unique")
```
